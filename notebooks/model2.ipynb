```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("user_fake_authentic_2class.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['class'] = le.fit_transform(df['class'])

print(df['class'].value_counts())
print(le.classes_)
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
for col in df.columns:
    if (df[col] < 0).any():
        print(col, df[col].min())
```

```python
duplicate_rows = df[df.duplicated(keep=False)]

print("Duplicated rows count:", len(duplicate_rows))
```

```python
print("Count of -1:", (df['cl'] == -1).sum())
print("Percentage:", ((df['cl'] == -1).sum()/len(df))*100)
```

```python
duplicate_rows = df[df.duplicated(keep=False)]

print(
    duplicate_rows.groupby(list(df.columns))
    .size()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
df['cl'] = df['cl'].replace(-1, 0)

df = df.drop_duplicates()
```

```python
print(df.shape)
print(df.duplicated().sum())
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('class', axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("\nTrain Labels:")
print(y_train.value_counts())

print("\nTest Labels:")
print(y_test.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall   :", recall_score(y_test, y_pred))
print("F1 Score :", f1_score(y_test, y_pred))
print("ROC AUC  :", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

```python
df = df.drop(columns=['pic'])
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('class', axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("\nTrain Labels:")
print(y_train.value_counts())

print("\nTest Labels:")
print(y_test.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf2 = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)

rf2.fit(X_train, y_train)
```

```python
train_pred2 = rf2.predict(X_train)
test_pred2 = rf2.predict(X_test)

from sklearn.metrics import accuracy_score

print("Train Accuracy:", accuracy_score(y_train, train_pred2))
print("Test Accuracy :", accuracy_score(y_test, test_pred2))
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix
)

y_pred2 = rf2.predict(X_test)
y_prob2 = rf2.predict_proba(X_test)[:,1]

print("Accuracy :", accuracy_score(y_test, y_pred2))
print("Precision:", precision_score(y_test, y_pred2))
print("Recall   :", recall_score(y_test, y_pred2))
print("F1 Score :", f1_score(y_test, y_pred2))
print("ROC AUC  :", roc_auc_score(y_test, y_prob2))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred2))
```
